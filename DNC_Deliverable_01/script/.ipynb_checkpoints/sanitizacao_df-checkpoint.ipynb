```python
import pandas as pd
pd.set_option('display.max_columns', None)
from pandas.api.types import CategoricalDtype

import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style()

from scipy import stats, linalg
```

```python
df = pd.read_csv('../dados/kc_house_data_pdata.csv', sep = ',')
```

### Excluir colunas que não serão utilizadas

```python
df = df.drop(columns=['exclur_1','excluir_2','excluir_3','cep'], axis=1)
df
```

```python
### Transformando coluna preço em BRl e convertendo valor em REAL
```

```python
#renomeando a coluna preço -> BRL
df = df.rename(columns = {'preço' : 'BRL'})

#convertendo colunas de números flutuantes para números inteiros e colocando 2 digitos após o ponto.
df[['banheiros', 'nº_pavtos']] = df[['banheiros', 'nº_pavtos']].apply(np.ceil).astype(int)
pd.options.display.float_format = "{:,.2f}".format

#convertendo valor de dolar em real (referênte ao dia 03/02/2022).
df['BRL'] = df['BRL'].apply(lambda x: x * 5.25)
df
```

```python
df.to_csv('../dados/kc_house_data_br.csv', sep = ',', index = False)
```

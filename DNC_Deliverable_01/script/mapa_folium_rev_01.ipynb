```python
import pandas as pd
pd.set_option('display.max_columns', None)
from pandas.api.types import CategoricalDtype

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style()

from scipy import stats, linalg

import folium
from folium.plugins import MarkerCluster
```

```python
df = pd.read_csv('../dados/kc_house_data.csv', parse_dates = ['date'])
df_type = CategoricalDtype(categories = df, ordered=True)
```

```python
df['view'] = df['view'].astype(df_type)
df['condition'] = df['condition'].astype(df_type)
df['grade'] = df['grade'].astype(CategoricalDtype(categories = df, ordered=False))
df['zipcode'] = df['zipcode'].astype(str)
df.sort_values(by=['date'])
df.head(50)
```

```python
map_h = folium.Map(location = [df['lat'].mean(), df['long'].mean()], zoom_start = 10)
marker_cluster = MarkerCluster().add_to(map_h)
MAX_RECORDS = 100

for name, row in df.iterrows():
    folium.Marker([row["lat"], row["long"]], 
                  popup="Sold for {0}$ on: {1}. Features: {2} sqft, {3} bedrooms, {4} bathrooms, year built: {5}".format(row["price"], row["date"], row['sqft_living'], row['bedrooms'], row['bathrooms'], row['yr_built'])).add_to(marker_cluster)

    
map_h.save('../maps/houses.html')
```
